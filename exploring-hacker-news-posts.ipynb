# Exploring Hacker News Posts

## Introduction
[Hacker News](https://news.ycombinator.com) is a platform initiated by the startup incubator Y Combinator, where user submitted stories, referred to as posts, receive votes and comments. The site is highly popular within technology and startup communities, and posts reaching the top listings can attract hundreds of thousands of visitors. This project explores two specific categories of posts: `'Ask HN'` and `'Show HN'`. `'Ask HN'` posts are submitted to pose a specific question to the Hacker News community. Conversely, `'Show HN'` posts are used to share a project, product, or general item of interest with the community.

## The Goal
The primary objective of this analysis is to compare these two post categories to determine the following:
Whether `'Ask HN'` or `'Show HN'` posts receive more comments on average.
Whether posts created at specific times receive more comments on average.

## Data Source
The [dataset](https://dq-content.s3.amazonaws.com/356/hacker_news.csv) utilised for this analysis contains approximately 20,000 rows. It was derived from an [original dataset](https://www.kaggle.com/datasets/hacker-news/hacker-news-posts) of nearly 300,000 rows by filtering out submissions with zero comments and subsequently taking a random sample from the remaining entries. 

The dataset comprises the following columns:
<br/>`id`: The unique identifier from Hacker News for the post.
<br/>`title`: The title of the post.
<br/>`url`: The URL that the post links to, if applicable.
<br/>`num_points`: The number of points the post acquired, calculated as the total number of upvotes less the total number of downvotes.
<br/>`num_comments`: The number of comments on the post.
<br/>`author`: The username of the person who submitted the post.
<br/>`created_at`: The date and time of the post submission.

## Read in the Data and Remove the Header

```python
import csv 
with open("Dataqust project - Hackers/hacker_news.csv") as file:
    hn = list(csv.reader(file))
```

```python
header = hn[0]
hn = hn[1:]

print(header)
print('\n')
print(hn[:5])
```

## Extracting Ask HN and Show HN Posts

```python
ask_posts = []
show_posts = []
other_posts = []

for post in hn:
    title = post[1]
    if title.lower().startswith("ask hn"):
        ask_posts.append(post)
    elif title.lower().startswith("show hn"):
        show_posts.append(post)
    else:
        other_posts.append(post)

print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
```

## Calculating the Average Number of Comments for Ask HN and Show HN Posts

```python
# Calculate the average number of comments `Ask HN` posts receive.
total_ask_comment = 0

for post in ask_posts:
    num_comment = int(post[4])
    total_ask_comment += num_comment

avg_ask_comment = total_ask_comment / len(ask_posts)
print(avg_ask_comment)
```

```python
# Calculate the average number of comments `Show HN` posts receive.
total_show_comment = 0

for post in show_posts:
    num_comment = int(post[4])
    total_show_comment += num_comment

avg_show_comment = total_show_comment / len(show_posts)
print(avg_show_comment)
```

On average, 'Ask HN' posts in the sample receive approximately 14 comments, whereas 'Show HN' posts receive approximately 10. Since 'Ask HN' posts are more likely to receive comments, the remaining analysis will focus exclusively on this category.

## Finding the Amount of Ask Posts and Comments by Hour Created



```python
import datetime as dt
result_list = []

for post in ask_posts:
    result_list.append([post[6], int(post[4])])

count_by_hour = {}
comments_by_hour = {}

for row in result_list:
    date = row[0]
    comment = row[1]
    time = dt.datetime.strptime(date,"%m/%d/%Y %H:%M")
    hour = time.strftime("%H")
    if hour in count_by_hour:
        comments_by_hour[hour] += comment
        count_by_hour[hour] += 1
    else:
        comments_by_hour[hour] = comment
        count_by_hour[hour] = 1

comments_by_hour
```

## Calculating the Average Number of Comments for Ask HN Posts by Hour

```python
avg_by_hour = []

for hour in comments_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] / count_by_hour[hour]])

avg_by_hour
```

## Sorting and Printing Values from a List of Lists

```python
# Invert the positions of the hour and the average comment value.
# Python sorts lists of lists by the first element of each sublist by default.
# Swapping the elements places the numerical average first, ensuring the 
# subsequent sorting is based on comment volume rather than chronological hours.
swap_avg_by_hour = []
for hour in avg_by_hour:
    swap_avg_by_hour.append([hour[1], hour[0]])

print(swap_avg_by_hour)
```

```python
sort_swap = sorted(swap_avg_by_hour, reverse = True)
print(sort_swap)
```

## Finding Top 5 Hours for Ask Posts Comments

```python
print("Top 5 Hour for Ask Posts Comments:")

for avg, hr in sort_swap[:5]:
    print(f"{dt.datetime.strptime(hr, "%H").strftime("%H:%M")} : {avg:2f} average comments per post")
    
    
```

The hour that receives the most comments per post on average is 15:00, with an average of 38.59 comments per post. There's about a 60% increase in the number of comments between the hours with the highest and second highest average number of comments.

## Conclusion

This project analysed a downsampled dataset of Hacker News submissions to determine whether 'Ask HN' or 'Show HN' posts receive more engagement, and to identify if the time of creation influences comment volume. 

The analysis revealed that `'Ask HN'` posts attract a higher volume of engagement, receiving an average of approximately 14.04 comments per post, compared to the 10.32 average comments received by `'Show HN'` posts. 

Furthermore, the hour at which a post is created heavily influences community interaction. For `'Ask HN'` posts, the **top five** hours for attracting comments are:

* **15:00 (3:00 PM EST)**: 38.59 average comments per post
* **02:00 (2:00 AM EST)**: 23.81 average comments per post
* **20:00 (8:00 PM EST)**: 21.53 average comments per post
* **16:00 (4:00 PM EST)**: 16.80 average comments per post
* **21:00 (9:00 PM EST)**: 16.01 average comments per post

The data indicates that 15:00 is the most effective time to submit an `'Ask HN'` post, yielding a significant increase in interaction compared to all other hours. Based on these findings, to maximise community engagement, it is recommended to categorise submissions as `'Ask HN'` and publish them during the mid-afternoon window, specifically around 15:00 EST.
